### P1. Write a Python code to output the following graphs. (or similar ones). 

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.arange(0,51)

plt.plot(x, 3*x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Draw a line')
plt.axis([0,50,0,160])

plt.show()
````

````python
x = np.arange(1.0, 3.5, 0.5)
y = np.array([2.0, 3, 4, 2.5, 1.0])

plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sample graph !')
plt.axis([1,3,1,4]),
plt.xticks(x, [1.0,1.5,2.0,2.5,3.0])
plt.show()
````

````python
x=[10,20,30]
y1=[20,40,10]
plt.plot(x, y1,label='line1-width-3',linewidth='3', color='b')
plt.plot(x,[40,10,30],label='line2-width-5',linewidth='5', color='r')


plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two or more lines with different widths and colors with suitable legends')
plt.axis([10,30,10,40]),
plt.xticks([10,15,20,25,30])
plt.legend()
plt.show()
````

````python
x=[10,20,30]
plt.plot(x, y1, 'b:',label='line1-width-3',linewidth='3')
plt.plot(x,[40,10,30],'r--',label='line2-width-5',linewidth='5')


plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two or more lines with different widths and colors with suitable legends')
plt.axis([10,30,10,40]),
plt.xticks([10,15,20,25,30])
plt.legend()
plt.show()
````

````python
x=[1,4,5,6,7]
y=[2,6,3,6,3]

plt.plot(x,y, 'r:',linewidth='2')
plt.scatter(x, y, c='blue')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Display marker')
plt.axis([1,8,1,8]),
plt.xticks(np.arange(1,9))
plt.show()
````

### P8. Plotly

```
Load the Iris dataset from plotly.express. 
```

````python
import plotly.express as px
df = px.data.iris()
df.head()
````

```
Create a scatter plot of petal_length vs petal_width. Use species as the color and display all the feature values when you hover over a point.
```

````python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", hover_data=['petal_length','petal_width','species_id'])
fig.show()
````

```
Create a histogram having 20 bins using the feature sepal_length. Use species feature as the color. 
```

````python
fig = px.histogram(df, x="sepal_length", nbins=20)
fig.show()
````

```
Plot sepal_length and sepal_width features on the same histogram. 
```

````python
fig = px.histogram(df, x=["sepal_length","sepal_width"], nbins=20)
fig.show()
````

```
Create a heatmap showing the correlation of the columns sepal_width and petal_width.
```

````python
df[['sepal_width','petal_width']]
print(df[['sepal_width','petal_width']].corr())
plt.matshow(df.corr())
plt.show()
````

```
Create a box plot showing the distribution of the sepal_length feature. 
```

````python
fig = px.box(df, x="sepal_length")
fig.show()
````
